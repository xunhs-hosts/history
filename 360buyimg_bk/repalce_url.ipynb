```python
from pathlib import Path
import pandas as pd 
import os, re
import random
import requests


root_dir = Path('./content')
dest_dir = Path('./temp')
```

```python
# _path = '/Users/only/Documents/githubs/hugoer/content/about.md'



query_str = r'\(https://m.360buyimg.com(.*)\)'



def download(url, _fp):
    fname = url.split('/')[-1]
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    r = requests.get(url, headers=headers)
    with open(_fp, 'wb') as fp:
        fp.write(r.content)
    return _fp

url_list = []
for _path in root_dir.glob('**/*.md'):
    with open(_path, 'r+') as fp:
        _str = fp.read()
        matched = re.findall(query_str, _str) 
        if matched != []:
            for m in matched:
                if m != '':
                    pre_url = 'https://img{}.360buyimg.com'.format(random.randint(10, 14))
                    full_url = pre_url + m
                    url_list.append(full_url)
```

```python
from tqdm.notebook import tqdm
for i in tqdm(url_list):
    fname_list = i.split('/')[-2:]
    fname = '_'.join(fname_list)
    _fp = dest_dir / fname
    if _fp.exists():
        print(_fp)
    else:
        print(i)
        download(i, _fp)
```
